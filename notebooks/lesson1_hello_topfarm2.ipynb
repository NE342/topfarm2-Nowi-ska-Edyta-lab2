```python
import numpy as np
import matplotlib.pyplot as plt

# Pozycje turbin
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

# Kolory dla każdej turbiny
colors = ['blue', 'red', 'green', 'yellow']

# Etykiety turbin
labels = ['T1', 'T2', 'T3', 'T4']

# Ustawienie figury
plt.figure(figsize=(8, 6))

# Rysowanie punktów z różnymi kolorami
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=colors[i], s=200, edgecolors='black', linewidth=2)
    # Dodanie etykiet dla każdej turbiny
    plt.text(x[i], y[i] + 50, labels[i], ha='center', fontsize=12, fontweight='bold')

# Formatowanie wykresu
plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami", fontsize=14, fontweight='bold')
plt.xlabel("x [m]", fontsize=12)
plt.ylabel("y [m]", fontsize=12)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Pozycje turbin (design variables)
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

# BOUNDARY - ograniczenia farmy
boundary_x = [0, 2000, 2000, 0, 0]      # prostokąt zamknięty
boundary_y = [0, 0, 2000, 2000, 0]

# Rysowanie
plt.figure(figsize=(10, 8))

# Rysuj granicę farmy
plt.plot(boundary_x, boundary_y, 'k-', linewidth=3, label='Granica farmy')
plt.fill(boundary_x, boundary_y, alpha=0.1, color='gray')

# Rysuj turbiny
plt.scatter(x, y, s=300, color='red', edgecolors='black', linewidth=2, label='Turbiny', zorder=5)

for i in range(len(x)):
    plt.text(x[i], y[i] + 80, f'T{i+1}', ha='center', fontsize=10, fontweight='bold')

# Formatowanie
plt.title("Farma wiatrowa - Boundary Constraints", fontsize=14, fontweight='bold')
plt.xlabel("x [m]", fontsize=12)
plt.ylabel("y [m]", fontsize=12)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.show()
```

## Ograniczenia farmy wiatrowej (Boundary Constraints)

**Co to jest?**
- Turbiny muszą być umieszczone w obszarze farmy (np. prostokąt 2000m x 2000m)
- To ograniczenie **boundary constraint** - definiuje granice działki

**Po co w praktyce?**
- Farma ma określone tereny (własność ziemi, zezwolenia)
- Nie możemy stawiać turbin poza tą linią
- Design variables (pozycje x, y) muszą spełniać: x ∈ [0, 2000], y ∈ [0, 2000]

**Przykład:** jeśli turbina wyjdzie poza granicę → solver to każe wciągnąć do wewnątrz
